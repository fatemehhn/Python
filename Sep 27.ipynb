```python
import numpy as np
```

```python
x = np.linspace(0,1,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
# functional method
```

```python
plt.plot(x,y,'r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
# multiplot in the same canvas
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
                
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(x,y)
axes2.set_title('Smaller Plot')
```
